Handling imbalanced datasets is crucial for predictive modeling since it can significantly bias the model towards the majority class, resulting in poor predictive performance on the minority class, which is often the class of interest.

### Addressing Imbalanced Classes

Several strategies can be employed to address imbalanced classes:

1.  **Resampling Techniques**:
    
    *   **Oversampling** the minority class or **undersampling** the majority class can help balance the classes. Techniques like SMOTE (Synthetic Minority Over-sampling Technique) can be used to synthetically generate new examples.
2.  **Altered Class Weights**:
    
    *   Some algorithms allow for weighting the classes differently to penalize misclassifications of the minority class more than those of the majority class.
3.  **Evaluation Metrics**:
    
    *   Using evaluation metrics that provide more insight into the performance on the minority class, such as precision, recall, F1-score (harmonic mean of precision and recall), and ROC-AUC, instead of relying solely on accuracy.

## 2. Feature Engineering and Selection

Multicollinearity occurs when two or more independent variables are highly correlated with one another in a regression model (which is the basis for some classification models like logistic regression).

Multicollinearity can be a problem in a regression model because we would not be able to distinguish between the individual effects of the independent variables on the dependent variable.

It may not affect the accuracy of the model as much, but we might lose reliability in determining the effects of individual features in our model which can be a problem in interpretability.

### Visualizing correlations between the numerical features

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

def plot_corr_heatmap(corr_matrix):
    # Create the heatmap trace
    heatmap_trace = go.Heatmap(z=corr_matrix.values,
                              x=corr_matrix.columns,
                              y=corr_matrix.index,
                              colorscale='RdBu_r',
                              colorbar=dict(title='Correlation'),
                              zmin=-1,  # Set the lower limit of the color scale
                              zmax=1)   # Set the upper limit of the color scale
    
    # Create the layout
    layout = go.Layout(title='Correlation Heatmap',
                       xaxis=dict(ticks='', tickangle=-45),
                       yaxis=dict(ticks=''),
                       hovermode='closest',
                       height=600)
    
    # Create the figure
    fig = go.Figure(data=[heatmap_trace], layout=layout)
    
    # Show the figure
    fig.show()

plot_corr_heatmap(corr_matrix)
```

### Visualizing strong feature correlations > 0.9
Some features are strongly correlated and can be dropped, to make the modelling faster and the models more robust and interpretable.

```python
r = corr_matrix
r[abs(r) <= 0.9] = 0
plot_corr_heatmap(r)
```

### Deciding which feature to drop based on lower variance
For each pair of feature with strong correlation > 90%, drop the feature with lower variance, because it contains less information.

```python
# Remove the upper half of the correlation matrix (excluding the diagonal)
r = r.where(np.triu(np.ones(r.shape), k=1).astype(bool))
# The NumPy triu function creates an upper triangular matrix of ones with the same shape as the correlation matrix.
# The k=1 parameter ensures that the diagonal and elements below it are excluded.
# The where method is then used to replace the values in the correlation matrix with NaN where the corresponding element in the upper triangular matrix is zero.

# Unstack the correlation matrix and sort the values to get the top correlations
top_correlations = r.unstack().sort_values(ascending=False)

# Get the top correlated pairs
highly_correlated_pairs = top_correlations[top_correlations > 0]
print(highly_correlated_pairs, '\n')

# Decide which feature to drop based on lower variance
features_to_drop = set()
for pair in highly_correlated_pairs.index:
    feature1, feature2 = pair
    variance1 = df[feature1].var()
    variance2 = df[feature2].var()
    
    # Choose the feature with lower variance to drop
    if variance1 < variance2:
        feature_to_exclude = feature1
    else:
        feature_to_exclude = feature2
    
    features_to_drop.add(feature_to_exclude)
    print(f"Exclude {feature_to_exclude} from the pair {pair}")
```

```python
print("These features will be dropped:")
features_to_drop
```

```python
df_reduced = df.drop(columns=features_to_drop)
df_reduced.head()
```

```python
# Reduced features:
plot_corr_heatmap(df_reduced.corr())
```
